# 1. import 불러오기 및 파일  불러오기

```python
import pandas as pd
import re
```

```python
def extract_and_combine_patterns(df, column_name, pattern):
    """ 주어진 패턴에 따라 DataFrame의 특정 열에서 데이터를 추출하고 결합합니다. """
    matches = df[column_name].fillna('').apply(lambda x: re.findall(pattern, x))
    combined = matches.apply(lambda x: ' '.join([' '.join(m[:2]) for m in x])).str.strip()
    return combined
```

# 4. 송수신공문 전처리(같은 날짜 데이터 합치기)

```python
import pandas as pd
# 수신 공문 추가
paper = pd.read_excel('/content/drive/MyDrive/11세부 구성기술3 데이터분류_230503/분류할 엑셀 공구별/2. 동해-삼척 3공구/5. 송수신공문/동해삼척_1공구 공문.xlsx', sheet_name="수신")
paper2 = pd.DataFrame(paper[['수신일자','문서명']])
paper2 ['수신일자'] = paper2['수신일자'].replace('[-]', '', regex=True)
paper2.rename(columns={'문서명':'문서명_수신'}, inplace=True)
paper2.rename(columns={'수신일자':'공사일지일자'}, inplace=True)
paper2
```

```python
# 날짜로 그룹화하고 데이터를 합침
paper2 = paper2.groupby('공사일지일자').sum().reset_index()
paper2
```

```python
# 발신 공문 추가
paper3 = pd.read_excel('/content/drive/MyDrive/11세부 구성기술3 데이터분류_230503/분류할 엑셀 공구별/2. 동해-삼척 3공구/5. 송수신공문/동해삼척_1공구 공문.xlsx', sheet_name="발신")
paper4 = pd.DataFrame(paper3[['시행일자','문서명']])
paper4 ['시행일자'] = paper4['시행일자'].replace('[-]', '', regex=True)
paper4.rename(columns={'문서명':'문서명_발신'}, inplace=True)
paper4.rename(columns={'시행일자':'공사일지일자'}, inplace=True)
paper4
```

```python
# 날짜로 그룹화하고 데이터를 합침
paper4 = paper4.groupby('공사일지일자').sum().reset_index()
paper4
```

```python
paper2['공사일지일자'] = paper2['공사일지일자'].astype(str)
paper4['공사일지일자'] = paper4['공사일지일자'].astype(str)
paper4 = pd.merge(paper2, paper4, how = 'left', on='공사일지일자')
paper4
```

```python
paper4.to_excel('/content/drive/MyDrive/11세부 구성기술3 데이터분류_230503/분류할 엑셀 공구별/2. 동해-삼척 3공구/4. 분류 완료 파일/동해-삼척 1공구 포장공사 분류 완료_수정_231123.xlsx',sheet_name = "포장공사분류")
```
