```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
#https://jimmy-ai.tistory.com/266
#d_type =  (), 'normal'  DB  
d_type = "chi2"
if d_type == "norm":
    mu = 7.4 # 
    sigma = 8.8 # 
    norm_dist = stats.norm(mu, sigma) #scipy norm  loc  .
    dist_lim = norm_dist.ppf(0.99)
    x1 = np.linspace(0, dist_lim, 1000) # x,  0 ,   99%,  1000 .
    y1 = norm_dist.pdf(x1)              #(pdf)  y
    y2 = norm_dist.cdf(x1)              #(cdf)  y
    x2 = norm_dist.ppf(0.95)            # 95% 
    fig_pdf = px.line(x = x1, y = y1)
    fig_cdf = px.line(x = x1, y = y2)
elif d_type == "gamma":
    mu = 2 # 
    sigma = 1.4 # 
    shape_parameter = round((mu ** 2)/(sigma ** 2),2)
    scale_parameter = round((mu)/(sigma ** 2),2) #scipy beta 
    scale_ = 1/scale_parameter #scipy  scale  1/beta
    gamma_dist = stats.gamma(shape_parameter, mu, scale_) #gamma(a, loc, scale)
    dist_lim = gamma_dist.ppf(0.99)
    x1 = np.linspace(0, dist_lim, 1000) # x,  0 ,   99%,  1000 .
    y1 = gamma_dist.pdf(x1)             #(pdf)  y
    y2 = gamma_dist.cdf(x1)             #(cdf)  y
    x2 = gamma_dist.ppf(0.95)           # 95% 
    fig_pdf = px.line(x = x1, y = y1)
    fig_cdf = px.line(x = x1, y = y2)
elif d_type == "expo":
    mu = 2 # 
    rate_parameter = round(1/mu,2) #scipy beta 
    scale_ = 1/rate_parameter #scipy  scale  1/beta
    expon_dist = stats.expon(mu, scale_) #expon(loc, scale)
    dist_lim = expon_dist.ppf(0.99)
    x1 = np.linspace(0, dist_lim, 1000) # x,  0 ,   99%,  1000 .
    y1 = expon_dist.pdf(x1)             #(pdf)  y
    y2 = expon_dist.cdf(x1)             #(cdf)  y
    x2 = expon_dist.ppf(0.95)           # 95% 
    fig_pdf = px.line(x = x1, y = y1)
    fig_cdf = px.line(x = x1, y = y2)
elif d_type == "chi2":
    df_ = 1 # param1
    chi2_dist = stats.chi2(df_) 
    dist_lim = chi2_dist.ppf(0.99)
    x1 = np.linspace(0, dist_lim, 1000) # x,  0 ,   99%,  1000 .
    y1 = chi2_dist.pdf(x1)              #(pdf)  y
    y2 = chi2_dist.cdf(x1)              #(cdf)  y
    x2 = chi2_dist.ppf(0.95)            # 95% 
    fig_pdf = px.line(x = x1, y = y1)
    fig_cdf = px.line(x = x1, y = y2)   
elif d_type == "t":
    mu = 1.83 # 
    sigma = 0.75 # 
    df_ = 5 # param1, r_p1
    t_dist = stats.t(df_, mu, sigma) #norm . loc=, scale=
    dist_lim = t_dist.ppf(0.99)
    x1 = np.linspace(0, dist_lim, 1000) # x,  0 ,   99%,  1000 .
    y1 = t_dist.pdf(x1)                 #(pdf)  y
    y2 = t_dist.cdf(x1)                 #(cdf)  y
    x2 = t_dist.ppf(0.95)               # 95% 
    #fig_pdf = go.Figure(go.Scatter(x=x1, y=y1, mode='lines')) #     
    fig_pdf = px.line(x = x1, y = y1)
    fig_cdf = px.line(x = x1, y = y2)

else:
    print("  .")

#  
fig_pdf.update_layout(title_text=" ")
fig_pdf.update_xaxes(title_text='())')
fig_pdf.update_yaxes(title_text=' ')
fig_pdf.update_layout(width = 600, height = 400, margin_r=50)
fig_cdf.update_layout(title_text="")
fig_cdf.update_xaxes(title_text='())')
fig_cdf.update_yaxes(title_text='')
fig_cdf.update_layout(width = 600, height = 400, margin_r=50)
#fig_cdf.add_hline(y = 0.95, line_color = "red")
#fig_cdf.add_vline(x = x2, line_color = "red")
fig_cdf.add_annotation(x = x2, y = 0.95, text = f" : {round(x2, 1)} ( 95%)", showarrow=True, arrowhead=6, arrowcolor="red")

fig_pdf.show()
fig_cdf.show()
x2
```
