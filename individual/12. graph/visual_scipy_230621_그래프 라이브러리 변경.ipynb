```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
#https://jimmy-ai.tistory.com/266
#d_type = 플랫폼 필드명(물리), 'normal' 추후 DB 연동 필요
d_type = "chi2"
if d_type == "norm":
    mu = 7.4 #데이터셋 평균
    sigma = 8.8 #데이터셋 표준편차
    norm_dist = stats.norm(mu, sigma) #scipy norm 함수에서 loc를 평균으로 정의함.
    dist_lim = norm_dist.ppf(0.99)
    x1 = np.linspace(0, dist_lim, 1000) #그래프 x축, 최소값 0 고정, 최대값 누적확률 99%, 사이즈 1000개 고정.
    y1 = norm_dist.pdf(x1)              #확률분포(pdf) 그래프 y축
    y2 = norm_dist.cdf(x1)              #누적확률분포(cdf) 그래프 y축
    x2 = norm_dist.ppf(0.95)            #누적확률 95%의 공기지연일
    fig_pdf = px.line(x = x1, y = y1)
    fig_cdf = px.line(x = x1, y = y2)
elif d_type == "gamma":
    mu = 2 #데이터셋 평균
    sigma = 1.4 #데이터셋 표준편차
    shape_parameter = round((mu ** 2)/(sigma ** 2),2)
    scale_parameter = round((mu)/(sigma ** 2),2) #scipy beta 값
    scale_ = 1/scale_parameter #scipy 내부 scale 변수는 1/beta
    gamma_dist = stats.gamma(shape_parameter, mu, scale_) #gamma(a, loc, scale)
    dist_lim = gamma_dist.ppf(0.99)
    x1 = np.linspace(0, dist_lim, 1000) #그래프 x축, 최소값 0 고정, 최대값 누적확률 99%, 사이즈 1000개 고정.
    y1 = gamma_dist.pdf(x1)             #확률분포(pdf) 그래프 y축
    y2 = gamma_dist.cdf(x1)             #누적확률분포(cdf) 그래프 y축
    x2 = gamma_dist.ppf(0.95)           #누적확률 95%의 공기지연일
    fig_pdf = px.line(x = x1, y = y1)
    fig_cdf = px.line(x = x1, y = y2)
elif d_type == "expo":
    mu = 2 #데이터셋 평균
    rate_parameter = round(1/mu,2) #scipy beta 값
    scale_ = 1/rate_parameter #scipy 내부 scale 변수는 1/beta
    expon_dist = stats.expon(mu, scale_) #expon(loc, scale)
    dist_lim = expon_dist.ppf(0.99)
    x1 = np.linspace(0, dist_lim, 1000) #그래프 x축, 최소값 0 고정, 최대값 누적확률 99%, 사이즈 1000개 고정.
    y1 = expon_dist.pdf(x1)             #확률분포(pdf) 그래프 y축
    y2 = expon_dist.cdf(x1)             #누적확률분포(cdf) 그래프 y축
    x2 = expon_dist.ppf(0.95)           #누적확률 95%의 공기지연일
    fig_pdf = px.line(x = x1, y = y1)
    fig_cdf = px.line(x = x1, y = y2)
elif d_type == "chi2":
    df_ = 1 #데이터셋 param1
    chi2_dist = stats.chi2(df_) 
    dist_lim = chi2_dist.ppf(0.99)
    x1 = np.linspace(0, dist_lim, 1000) #그래프 x축, 최소값 0 고정, 최대값 누적확률 99%, 사이즈 1000개 고정.
    y1 = chi2_dist.pdf(x1)              #확률분포(pdf) 그래프 y축
    y2 = chi2_dist.cdf(x1)              #누적확률분포(cdf) 그래프 y축
    x2 = chi2_dist.ppf(0.95)            #누적확률 95%의 공기지연일
    fig_pdf = px.line(x = x1, y = y1)
    fig_cdf = px.line(x = x1, y = y2)   
elif d_type == "t":
    mu = 1.83 #데이터셋 평균
    sigma = 0.75 #데이터셋 표준편차
    df_ = 5 #데이터셋 param1, r_p1
    t_dist = stats.t(df_, mu, sigma) #norm에서 확장됨. loc=기대값, scale=표준편차
    dist_lim = t_dist.ppf(0.99)
    x1 = np.linspace(0, dist_lim, 1000) #그래프 x축, 최소값 0 고정, 최대값 누적확률 99%, 사이즈 1000개 고정.
    y1 = t_dist.pdf(x1)                 #확률분포(pdf) 그래프 y축
    y2 = t_dist.cdf(x1)                 #누적확률분포(cdf) 그래프 y축
    x2 = t_dist.ppf(0.95)               #누적확률 95%의 공기지연일
    #fig_pdf = go.Figure(go.Scatter(x=x1, y=y1, mode='lines')) #추후 업데이트 위한 그래프 생성 코드
    fig_pdf = px.line(x = x1, y = y1)
    fig_cdf = px.line(x = x1, y = y2)

else:
    print("일치하는 확률분포가 없습니다.")

#그래프 디자인 항목
fig_pdf.update_layout(title_text="공기지연 확률분포")
fig_pdf.update_xaxes(title_text='공기지연(일))')
fig_pdf.update_yaxes(title_text=' ')
fig_pdf.update_layout(width = 600, height = 400, margin_r=50)
fig_cdf.update_layout(title_text="누적확률분포")
fig_cdf.update_xaxes(title_text='공기지연(일))')
fig_cdf.update_yaxes(title_text='누적확률')
fig_cdf.update_layout(width = 600, height = 400, margin_r=50)
#fig_cdf.add_hline(y = 0.95, line_color = "red")
#fig_cdf.add_vline(x = x2, line_color = "red")
fig_cdf.add_annotation(x = x2, y = 0.95, text = f"공기지연 : {round(x2, 1)}일 (누적확률 95%)", showarrow=True, arrowhead=6, arrowcolor="red")

fig_pdf.show()
fig_cdf.show()
x2
```
