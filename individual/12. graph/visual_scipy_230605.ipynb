```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
#https://jimmy-ai.tistory.com/266
#d_type = 플랫폼 필드명(물리), 'normal' 추후 DB 연동 필요
d_type = "gamma"
if d_type == "norm":
    mu = 7.4 #데이터셋 평균
    sigma = 8.8 #데이터셋 표준편차
    norm_dist = stats.norm(mu, sigma) #scipy norm 함수에서 loc를 평균으로 정의함.
    dist_lim = norm_dist.ppf(0.99)
    x = np.linspace(0, dist_lim, 1000) #최소값 0 고정, 최대값 누적확률 99%, 사이즈 1000개 고정.
    y1 = norm_dist.pdf(x)                #확률분포
    plt.figure(figsize=(10, 6))          # 플롯 사이즈 지정
    plt.plot(x, y1, color="red")         # 선을 빨강색으로 지정하여 plot 작성          
    plt.xlabel("x")                      # x축 레이블 지정
    plt.ylabel("y1")                      # y축 레이블 지정
    plt.grid()                           # 플롯에 격자 보이기
    plt.title("Normal Distribution with scipy.stats")     # 타이틀 표시
    plt.legend(["N(" + str(mu) + ", " + str(sigma) + ")"])  # 범례 표시
    plt.show()
    y2 = norm_dist.cdf(x)                #누적확률분포
    y3 = norm_dist.ppf(0.95)             #누적확률분포 구간
    plt.figure(figsize=(10, 6))          # 플롯 사이즈 지정
    plt.plot(x, y2, color="red")         # 선을 빨강색으로 지정하여 plot 작성          
    plt.xlabel("x")                      # x축 레이블 지정
    plt.ylabel("y2")                      # y축 레이블 지정
    plt.grid()                           # 플롯에 격자 보이기
    plt.title("cumul Distribution with scipy.stats")     # 타이틀 표시
    plt.legend(["N(" + str(mu) + ", " + str(sigma) + ")"])  # 범례 표시
    plt.scatter(y3, 0.95, c = 'r') # p = 0.95인점 표시
    plt.plot([0, y3], [0.95, 0.95], 'r--')
    plt.plot([y3, y3], [0.95, 0], 'r--')
    plt.show() 
    print(y3)
elif d_type == "gamma":
    mu = 2 #데이터셋 평균
    sigma = 1.4 #데이터셋 표준편차
    shape_parameter = round((mu ** 2)/(sigma ** 2),2)
    scale_parameter = round((mu)/(sigma ** 2),2) #scipy beta 값
    scale_ = 1/scale_parameter #scipy 내부 scale 변수는 1/beta
    gamma_dist = stats.gamma(shape_parameter, mu, scale_) #gamma(a, loc, scale)
    dist_lim = gamma_dist.ppf(0.99)
    x = np.linspace(0, dist_lim, 1000) #최소값 0 고정, 최대값 누적확률 99%, 사이즈 1000개 고정.
    y1 = gamma_dist.pdf(x)                #확률분포
    plt.figure(figsize=(10, 6))          # 플롯 사이즈 지정
    plt.plot(x, y1, color="red")         # 선을 빨강색으로 지정하여 plot 작성          
    plt.xlabel("x")                      # x축 레이블 지정
    plt.ylabel("y1")                      # y축 레이블 지정
    plt.grid()                           # 플롯에 격자 보이기
    plt.title("Gamma Distribution with scipy.stats")     # 타이틀 표시
    plt.legend(["Shape(" + str(shape_parameter) + ", scale" + str(scale_parameter) + ")"])  # 범례 표시
    plt.show()
    y2 = gamma_dist.cdf(x)                #누적확률분포
    y3 = gamma_dist.ppf(0.95)             #누적확률분포 구간
    plt.figure(figsize=(10, 6))          # 플롯 사이즈 지정
    plt.plot(x, y2, color="red")         # 선을 빨강색으로 지정하여 plot 작성          
    plt.xlabel("x")                      # x축 레이블 지정
    plt.ylabel("y2")                      # y축 레이블 지정
    plt.grid()                           # 플롯에 격자 보이기
    plt.title("cumul Distribution with scipy.stats")     # 타이틀 표시
    plt.legend(["Shape(" + str(shape_parameter) + ", scale" + str(scale_parameter) + ")"])  # 범례 표시
    plt.scatter(y3, 0.95, c = 'r') # p = 0.95인점 표시
    plt.plot([0, y3], [0.95, 0.95], 'r--')
    plt.plot([y3, y3], [0.95, 0], 'r--')
    plt.show() 
    print(y3)
elif d_type == "expo":
    mu = 2 #데이터셋 평균
    rate_parameter = round(1/mu,2) #scipy beta 값
    scale_ = 1/rate_parameter #scipy 내부 scale 변수는 1/beta
    expon_dist = stats.expon(mu, scale_) #expon(loc, scale)
    dist_lim = expon_dist.ppf(0.99)
    x = np.linspace(0, dist_lim, 1000) #최소값 0 고정, 최대값 누적확률 99%, 사이즈 1000개 고정.
    y1 = expon_dist.pdf(x)                #확률분포
    plt.figure(figsize=(10, 6))          # 플롯 사이즈 지정
    plt.plot(x, y1, color="red")         # 선을 빨강색으로 지정하여 plot 작성          
    plt.xlabel("x")                      # x축 레이블 지정
    plt.ylabel("y1")                      # y축 레이블 지정
    plt.grid()                           # 플롯에 격자 보이기
    plt.title("Gamma Distribution with scipy.stats")     # 타이틀 표시
    plt.legend(["rate(" + str(rate_parameter) + ")"])  # 범례 표시
    plt.show()
    y2 = expon_dist.cdf(x)                #누적확률분포
    y3 = expon_dist.ppf(0.95)             #누적확률분포 구간
    plt.figure(figsize=(10, 6))          # 플롯 사이즈 지정
    plt.plot(x, y2, color="red")         # 선을 빨강색으로 지정하여 plot 작성          
    plt.xlabel("x")                      # x축 레이블 지정
    plt.ylabel("y2")                      # y축 레이블 지정
    plt.grid()                           # 플롯에 격자 보이기
    plt.title("cumul Distribution with scipy.stats")     # 타이틀 표시
    plt.legend(["rate(" + str(rate_parameter) + ")"])  # 범례 표시
    plt.scatter(y3, 0.95, c = 'r') # p = 0.95인점 표시
    plt.plot([0, y3], [0.95, 0.95], 'r--')
    plt.plot([y3, y3], [0.95, 0], 'r--')
    plt.show() 
    print(y3)
elif d_type == "chi2":
    df_ = 1 #데이터셋 param1
    chi2_dist = stats.chi2(df_) 
    dist_lim = chi2_dist.ppf(0.99)
    x = np.linspace(0, dist_lim, 1000) #최소값 0 고정, 최대값 누적확률 99%, 사이즈 1000개 고정.
    y1 = chi2_dist.pdf(x)                #확률분포
    plt.figure(figsize=(10, 6))          # 플롯 사이즈 지정
    plt.plot(x, y1, color="red")         # 선을 빨강색으로 지정하여 plot 작성          
    plt.xlabel("x")                      # x축 레이블 지정
    plt.ylabel("y1")                      # y축 레이블 지정
    plt.grid()                           # 플롯에 격자 보이기
    plt.title("Gamma Distribution with scipy.stats")     # 타이틀 표시
    plt.legend(["df(" + str(df_) + ")"])  # 범례 표시
    plt.show()
    y2 = chi2_dist.cdf(x)                #누적확률분포
    y3 = chi2_dist.ppf(0.95)             #누적확률분포 구간
    plt.figure(figsize=(10, 6))          # 플롯 사이즈 지정
    plt.plot(x, y2, color="red")         # 선을 빨강색으로 지정하여 plot 작성          
    plt.xlabel("x")                      # x축 레이블 지정
    plt.ylabel("y2")                      # y축 레이블 지정
    plt.grid()                           # 플롯에 격자 보이기
    plt.title("cumul Distribution with scipy.stats")     # 타이틀 표시
    plt.legend(["df(" + str(df_) + ")"])  # 범례 표시
    plt.scatter(y3, 0.95, c = 'r') # p = 0.95인점 표시
    plt.plot([0, y3], [0.95, 0.95], 'r--')
    plt.plot([y3, y3], [0.95, 0], 'r--')
    plt.show() 
    print(y3)    
elif d_type == "t":
    mu = 1.83 #데이터셋 평균
    sigma = 0.75 #데이터셋 표준편차
    df_ = 5 #데이터셋 param1, r_p1
    t_dist = stats.t(df_, mu, sigma) #norm에서 확장됨. loc=기대값, scale=표준편차
    dist_lim = t_dist.ppf(0.99)
    x = np.linspace(0, dist_lim, 1000) #최소값 0 고정, 최대값 누적확률 99%, 사이즈 1000개 고정.
    y1 = t_dist.pdf(x)                #확률분포
    plt.figure(figsize=(10, 6))          # 플롯 사이즈 지정
    plt.plot(x, y1, color="red")         # 선을 빨강색으로 지정하여 plot 작성          
    plt.xlabel("x")                      # x축 레이블 지정
    plt.ylabel("y1")                      # y축 레이블 지정
    plt.grid()                           # 플롯에 격자 보이기
    plt.title("t Distribution with scipy.stats")     # 타이틀 표시
    plt.legend(["df(" + str(df_) + ")," + "N(" + str(mu) + ", " + str(sigma) + ")"])  # 범례 표시
    plt.show()
    y2 = t_dist.cdf(x)                #누적확률분포
    y3 = t_dist.ppf(0.95)             #누적확률분포 구간
    plt.figure(figsize=(10, 6))          # 플롯 사이즈 지정
    plt.plot(x, y2, color="red")         # 선을 빨강색으로 지정하여 plot 작성          
    plt.xlabel("x")                      # x축 레이블 지정
    plt.ylabel("y2")                      # y축 레이블 지정
    plt.grid()                           # 플롯에 격자 보이기
    plt.title("t Distribution with scipy.stats")     # 타이틀 표시
    plt.legend(["df(" + str(df_) + ")," + "N(" + str(mu) + ", " + str(sigma) + ")"])  # 범례 표시
    plt.scatter(y3, 0.95, c = 'r') # p = 0.95인점 표시
    plt.plot([0, y3], [0.95, 0.95], 'r--')
    plt.plot([y3, y3], [0.95, 0], 'r--')
    plt.show() 
    print(y3)
else:
    print("일치하는 확률분포가 없습니다.")    
```
