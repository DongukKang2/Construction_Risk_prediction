```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
#https://jimmy-ai.tistory.com/266
#d_type =  (), 'normal'  DB  
d_type = "gamma"
if d_type == "norm":
    mu = 7.4 # 
    sigma = 8.8 # 
    norm_dist = stats.norm(mu, sigma) #scipy norm  loc  .
    dist_lim = norm_dist.ppf(0.99)
    x = np.linspace(0, dist_lim, 1000) # 0 ,   99%,  1000 .
    y1 = norm_dist.pdf(x)                #
    plt.figure(figsize=(10, 6))          #   
    plt.plot(x, y1, color="red")         #    plot           
    plt.xlabel("x")                      # x  
    plt.ylabel("y1")                      # y  
    plt.grid()                           #   
    plt.title("Normal Distribution with scipy.stats")     #  
    plt.legend(["N(" + str(mu) + ", " + str(sigma) + ")"])  #  
    plt.show()
    y2 = norm_dist.cdf(x)                #
    y3 = norm_dist.ppf(0.95)             # 
    plt.figure(figsize=(10, 6))          #   
    plt.plot(x, y2, color="red")         #    plot           
    plt.xlabel("x")                      # x  
    plt.ylabel("y2")                      # y  
    plt.grid()                           #   
    plt.title("cumul Distribution with scipy.stats")     #  
    plt.legend(["N(" + str(mu) + ", " + str(sigma) + ")"])  #  
    plt.scatter(y3, 0.95, c = 'r') # p = 0.95 
    plt.plot([0, y3], [0.95, 0.95], 'r--')
    plt.plot([y3, y3], [0.95, 0], 'r--')
    plt.show() 
    print(y3)
elif d_type == "gamma":
    mu = 2 # 
    sigma = 1.4 # 
    shape_parameter = round((mu ** 2)/(sigma ** 2),2)
    scale_parameter = round((mu)/(sigma ** 2),2) #scipy beta 
    scale_ = 1/scale_parameter #scipy  scale  1/beta
    gamma_dist = stats.gamma(shape_parameter, mu, scale_) #gamma(a, loc, scale)
    dist_lim = gamma_dist.ppf(0.99)
    x = np.linspace(0, dist_lim, 1000) # 0 ,   99%,  1000 .
    y1 = gamma_dist.pdf(x)                #
    plt.figure(figsize=(10, 6))          #   
    plt.plot(x, y1, color="red")         #    plot           
    plt.xlabel("x")                      # x  
    plt.ylabel("y1")                      # y  
    plt.grid()                           #   
    plt.title("Gamma Distribution with scipy.stats")     #  
    plt.legend(["Shape(" + str(shape_parameter) + ", scale" + str(scale_parameter) + ")"])  #  
    plt.show()
    y2 = gamma_dist.cdf(x)                #
    y3 = gamma_dist.ppf(0.95)             # 
    plt.figure(figsize=(10, 6))          #   
    plt.plot(x, y2, color="red")         #    plot           
    plt.xlabel("x")                      # x  
    plt.ylabel("y2")                      # y  
    plt.grid()                           #   
    plt.title("cumul Distribution with scipy.stats")     #  
    plt.legend(["Shape(" + str(shape_parameter) + ", scale" + str(scale_parameter) + ")"])  #  
    plt.scatter(y3, 0.95, c = 'r') # p = 0.95 
    plt.plot([0, y3], [0.95, 0.95], 'r--')
    plt.plot([y3, y3], [0.95, 0], 'r--')
    plt.show() 
    print(y3)
elif d_type == "expo":
    mu = 2 # 
    rate_parameter = round(1/mu,2) #scipy beta 
    scale_ = 1/rate_parameter #scipy  scale  1/beta
    expon_dist = stats.expon(mu, scale_) #expon(loc, scale)
    dist_lim = expon_dist.ppf(0.99)
    x = np.linspace(0, dist_lim, 1000) # 0 ,   99%,  1000 .
    y1 = expon_dist.pdf(x)                #
    plt.figure(figsize=(10, 6))          #   
    plt.plot(x, y1, color="red")         #    plot           
    plt.xlabel("x")                      # x  
    plt.ylabel("y1")                      # y  
    plt.grid()                           #   
    plt.title("Gamma Distribution with scipy.stats")     #  
    plt.legend(["rate(" + str(rate_parameter) + ")"])  #  
    plt.show()
    y2 = expon_dist.cdf(x)                #
    y3 = expon_dist.ppf(0.95)             # 
    plt.figure(figsize=(10, 6))          #   
    plt.plot(x, y2, color="red")         #    plot           
    plt.xlabel("x")                      # x  
    plt.ylabel("y2")                      # y  
    plt.grid()                           #   
    plt.title("cumul Distribution with scipy.stats")     #  
    plt.legend(["rate(" + str(rate_parameter) + ")"])  #  
    plt.scatter(y3, 0.95, c = 'r') # p = 0.95 
    plt.plot([0, y3], [0.95, 0.95], 'r--')
    plt.plot([y3, y3], [0.95, 0], 'r--')
    plt.show() 
    print(y3)
elif d_type == "chi2":
    df_ = 1 # param1
    chi2_dist = stats.chi2(df_) 
    dist_lim = chi2_dist.ppf(0.99)
    x = np.linspace(0, dist_lim, 1000) # 0 ,   99%,  1000 .
    y1 = chi2_dist.pdf(x)                #
    plt.figure(figsize=(10, 6))          #   
    plt.plot(x, y1, color="red")         #    plot           
    plt.xlabel("x")                      # x  
    plt.ylabel("y1")                      # y  
    plt.grid()                           #   
    plt.title("Gamma Distribution with scipy.stats")     #  
    plt.legend(["df(" + str(df_) + ")"])  #  
    plt.show()
    y2 = chi2_dist.cdf(x)                #
    y3 = chi2_dist.ppf(0.95)             # 
    plt.figure(figsize=(10, 6))          #   
    plt.plot(x, y2, color="red")         #    plot           
    plt.xlabel("x")                      # x  
    plt.ylabel("y2")                      # y  
    plt.grid()                           #   
    plt.title("cumul Distribution with scipy.stats")     #  
    plt.legend(["df(" + str(df_) + ")"])  #  
    plt.scatter(y3, 0.95, c = 'r') # p = 0.95 
    plt.plot([0, y3], [0.95, 0.95], 'r--')
    plt.plot([y3, y3], [0.95, 0], 'r--')
    plt.show() 
    print(y3)    
elif d_type == "t":
    mu = 1.83 # 
    sigma = 0.75 # 
    df_ = 5 # param1, r_p1
    t_dist = stats.t(df_, mu, sigma) #norm . loc=, scale=
    dist_lim = t_dist.ppf(0.99)
    x = np.linspace(0, dist_lim, 1000) # 0 ,   99%,  1000 .
    y1 = t_dist.pdf(x)                #
    plt.figure(figsize=(10, 6))          #   
    plt.plot(x, y1, color="red")         #    plot           
    plt.xlabel("x")                      # x  
    plt.ylabel("y1")                      # y  
    plt.grid()                           #   
    plt.title("t Distribution with scipy.stats")     #  
    plt.legend(["df(" + str(df_) + ")," + "N(" + str(mu) + ", " + str(sigma) + ")"])  #  
    plt.show()
    y2 = t_dist.cdf(x)                #
    y3 = t_dist.ppf(0.95)             # 
    plt.figure(figsize=(10, 6))          #   
    plt.plot(x, y2, color="red")         #    plot           
    plt.xlabel("x")                      # x  
    plt.ylabel("y2")                      # y  
    plt.grid()                           #   
    plt.title("t Distribution with scipy.stats")     #  
    plt.legend(["df(" + str(df_) + ")," + "N(" + str(mu) + ", " + str(sigma) + ")"])  #  
    plt.scatter(y3, 0.95, c = 'r') # p = 0.95 
    plt.plot([0, y3], [0.95, 0.95], 'r--')
    plt.plot([y3, y3], [0.95, 0], 'r--')
    plt.show() 
    print(y3)
else:
    print("  .")    
```
